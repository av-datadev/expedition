Does a bigger battery mean a longer range?

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='battery_capacity_kWh',y='range_km',hue='brand',data=df)
plt.title('Battery Capacity vs Range')
plt.xlabel('Battery Capacity')
plt.ylabel('Range')
plt.grid(True)
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

Does higher top speed come with slower acceleration?

```python
sns.scatterplot(x='top_speed_kmh',y='acceleration_0_100_s',data=df,hue='brand')
plt.title('Top Speed vs Acc')
plt.xlabel('Top Speed')
plt.ylabel('Accleration')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

Which drivetrain is more efficient?

```python
sns.barplot(x='drivetrain',y='range_km',data=df)
plt.title('Efficiency for drivetrain')
plt.xlabel('Drive Train')
plt.ylabel('Range')
plt.show()
```

Does cargo_volume_l vary by car_body_type?

```python
plt.figure(figsize=(15,8))
sns.barplot(y='cargo_volume_l',x='car_body_type',data=df)
plt.title('Cargo Vol vs Car Body Type')
plt.xlabel('Car body Type')
plt.ylabel('Cargo Volume')
plt.show()
```

Create new categorical columns based on numerical ranges (e.g., fast vs slow acceleration).

```python
df['acceleration_category'] = pd.cut(df['acceleration_0_100_s'],
                                    bins=[0,5,8,float('inf')],
                                    labels=['Fast','Average','Slow']
                                    )
```
